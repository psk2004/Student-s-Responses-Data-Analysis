```python
#Correlation b/w the time spent on self study and the same for social media

corr_df = df_student.drop(["Time spent on self study","Time spent on social media"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df,annot = True,fmt = '.1g', vmin = -1, vmax = 1, center = 0, cmap = "inferno", linewidth = 1, linecolor ="Black")
heatmap.set_title("Correlation b/w the time spent on self study and the same for social media")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation = 90)
```

```python
rating_counts = df_student['Rating of Online Class experience'].value_counts()              #Graph to show the ratings of online class experience

# Sort the ratings by index (which is the rating category)
rating_counts_sorted = rating_counts.sort_index()

# Plotting the bar graph
plt.figure(figsize=(8, 6))
rating_counts_sorted.plot(kind='bar', color='skyblue')
plt.title('Distribution of Ratings of Online Class Experience')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
weight_change_counts = df_student['Change in your weight'].value_counts()

# Sort the weight change categories by index
weight_change_counts_sorted = weight_change_counts.sort_index()

# Plotting the bar graph
plt.figure(figsize=(8, 6))
weight_change_counts_sorted.plot(kind='bar', color='lightgreen')
plt.title('Change in Weight During Lockdown')
plt.xlabel('Change in Weight')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
social_media_counts = df_student['Prefered social media platform'].value_counts()                          #Preferred social media platforms for students using pie chart

plt.figure(figsize=(8, 6))
plt.pie(social_media_counts, labels=social_media_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Preferred Social Media Platforms')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
#Displaying the time students spent on social media and the stress busters they used; sorted on the time spent on social media. We want to check if social media has anything to do with stress
sorted_df = df_student.groupby(["Time spent on social media"])["Stress busters"].max().reset_index().sort_values(ascending = False,by = "Time spent on social media")                 #We are displaying the time a person had spent their day while studying or by seeing TV.
sorted_df
```

```python
health_issue_yes = df_student[df_student['Health issue during lockdown'] == 'YES']['Time spent on fitness']
health_issue_no = df_student[df_student['Health issue during lockdown'] == 'NO']['Time spent on fitness']

# Box plot to compare the distribution of time spent on fitness between students who experienced health issues during lockdown and those who did not.
plt.figure(figsize=(8, 6))
sns.boxplot(x='Health issue during lockdown', y='Time spent on fitness', data=df_student)
plt.title('Time Spent on Fitness by Health Issue During Lockdown')
plt.xlabel('Health Issue During Lockdown')
plt.ylabel('Time Spent on Fitness')
plt.show()
```
